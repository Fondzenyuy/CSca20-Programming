```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar' , 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov','Dec']
temperatures = [0,10,2.5,5,15,20,22.5,25,14.5,5.5,2.5,0]

plt.plot(months, temperatures)
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Prelab output')
plt.show()
```

```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar' , 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov','Dec']
temp = []
start_year= input("Please select start year")
end_year= input('end year')
a = int(start_year)
b = int(end_year)

for current_year in range(a,b+1,1):
    temp = []
    for current_month in months:
        month_temp = int(input("Please enter average temperature in" + str(current_month) + 'of' + str(current_year) + "."))
        temp = temp + [month_temp]
    plt.plot(months,temp)

plt.xlabel('Months')
plt.ylabel('Temperature')
plt.show()  
```
